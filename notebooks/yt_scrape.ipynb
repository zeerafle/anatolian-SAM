```python
import yt_dlp
import os
```

```python
# data location
DATA_DIR_RAW = "../data/raw"
DATA_DIR_BAGLAMA = os.path.join(DATA_DIR_RAW, "baglama")
DATA_DIR_ZURNA = os.path.join(DATA_DIR_RAW, "zurna")

if not os.path.exists(DATA_DIR_BAGLAMA):
    os.makedirs(DATA_DIR_BAGLAMA)
if not os.path.exists(DATA_DIR_ZURNA):
    os.makedirs(DATA_DIR_ZURNA)
```

```python
URLS_baglama = ['https://www.youtube.com/playlist?list=PLp0lDtdH8F_FHnUKfVOPkUD3pgQcuVxgd']

ydl_opts = {
    'format': 'wav/bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'wav',
    }],
    # save to data directory with id and title as filename
    'outtmpl': os.path.join(DATA_DIR_BAGLAMA, '%(id)s-%(title)s.%(ext)s'),
    'ignoreerrors': True,
    'download_archive': '../downloaded_videos.txt',
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    error_code = ydl.download(URLS_baglama)
```

```python
URLS_zurna = [
    'https://www.youtube.com/watch?v=YEd4i5phVBo',
]

ydl_opts = {
    'format': 'wav/bestaudio/best',
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'wav',
    }],
    # save to data directory with id and title as filename
    'outtmpl': os.path.join(DATA_DIR_ZURNA, '%(id)s-%(title)s.%(ext)s'),
    'ignoreerrors': True,
    'download_archive': '../downloaded_videos.txt',
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    error_code = ydl.download(URLS_zurna)
```
